```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import re
import glob
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# ========== КОНФИГУРАЦИЯ ==========
# Пути к папкам с данными
data_folders = {
    'X': r'C:\Users\VKosinov\Desktop\GitHub\work\acceleration_error\data_X',
    'Y': r'C:\Users\VKosinov\Desktop\GitHub\work\acceleration_error\data_Y',
    'Z': r'C:\Users\VKosinov\Desktop\GitHub\work\acceleration_error\data_Z'
}

# Путь для общего финального файла
final_all_path = r'C:\Users\VKosinov\Desktop\GitHub\work\acceleration_error\final.csv'

# ========== 0. ПРЕДВАРИТЕЛЬНАЯ ОБРАБОТКА ФАЙЛОВ ==========
def extract_info_from_filename(filename):
    """
    Извлекает информацию из имени файла формата: 
    примеры: 1-s 5hz-0.1.csv, 2-x 10hz-5.6.csv, 3-s 20hz-3.2.csv
    Возвращает: sensor, source_file, frequency, boost
    """
    # Удаляем расширение .csv
    base_name = filename.replace('.csv', '').lower()
    
    # Инициализируем значения по умолчанию
    sensor = None
    source_file = None
    frequency = None
    boost = None
    
    try:
        # Разделяем по пробелам и дефисам
        parts = re.split(r'[\s\-]+', base_name)
        
        # sensor - первый элемент (цифра)
        if parts and parts[0].isdigit():
            sensor = parts[0]
        
        # source_file - ищем 's' или 'x'/'y'/'z'
        for part in parts:
            if 's' in part.lower() and all(axis not in part.lower() for axis in ['x', 'y', 'z']):
                source_file = 'S'
                break
            elif 'x' in part.lower() and 's' not in part.lower():
                source_file = 'X'
                break
            elif 'y' in part.lower() and 's' not in part.lower():
                source_file = 'Y'
                break
            elif 'z' in part.lower() and 's' not in part.lower():
                source_file = 'Z'
                break
        
        # frequency - ищем значение с 'hz'
        for part in parts:
            if 'hz' in part.lower():
                # Извлекаем только цифры
                freq_num = re.findall(r'\d+', part)
                if freq_num:
                    frequency = freq_num[0]
                break
        
        # boost - последний элемент (после последнего дефиса)
        for i in range(len(parts)-1, -1, -1):
            # Проверяем, является ли часть числом (может быть с точкой)
            if re.match(r'^\d+(\.\d+)?$', parts[i]):
                boost = parts[i]
                break
                
    except Exception as e:
        print(f"Ошибка при разборе имени файла {filename}: {e}")
    
    return sensor, source_file, frequency, boost

def format_number(x):
    try:
        return f"{float(x):.1f}".replace('.', ',')
    except:
        return x

# Список для сбора всех данных
all_sensors_data = []

# Обрабатываем данные для каждой оси
for axis, folder_path in data_folders.items():
    print(f"\n{'='*80}")
    print(f"ОБРАБОТКА ОСИ {axis}")
    print(f"{'='*80}")
    
    # Находим все CSV файлы в папке (кроме final.csv)
    csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
    csv_files = [f for f in csv_files if "final.csv" not in f.lower()]
    
    print(f"Найдено {len(csv_files)} CSV файлов для обработки")
    
    # Обрабатываем каждый файл перед основным анализом
    all_mean_rows = []
    processed_files_count = 0
    
    for i, file_path in enumerate(csv_files, 1):
        filename = os.path.basename(file_path)
        
        try:
            # Извлекаем информацию из имени файла
            sensor_val, source_file_val, frequency_val, boost_val = extract_info_from_filename(filename)
            
            # ФИЛЬТР: оставляем только строки с 'S'
            if source_file_val != 'S':
                print(f"  Пропускаем файл {filename}: source_file = {source_file_val} (не 'S')")
                continue
                
            if not all([sensor_val, source_file_val, frequency_val, boost_val]):
                print(f"  Внимание: не все данные извлечены из имени файла")
                print(f"  Извлечено: sensor={sensor_val}, source_file={source_file_val}, frequency={frequency_val}, boost={boost_val}")
            
            # Читаем данные - только 2 столбца без имен
            df = pd.read_csv(file_path, header=None, delimiter=';', encoding='cp1251')
            
            # Оставляем только второй столбец и переименовываем его
            if df.shape[1] >= 2:
                df = df[[1]]  # Второй столбец (индекс 1)
                df.columns = ['acceleration']
            elif df.shape[1] == 1:
                # Если только один столбец, используем его
                df.columns = ['acceleration']
            else:
                print(f"  Внимание: неожиданное количество столбцов: {df.shape[1]}")
                continue
            
            # УДАЛЕНИЕ ПРОПУСКОВ
            initial_count = len(df)
            df = df.dropna()  # Удаляем строки с NaN значениями
            df = df.replace('', np.nan).dropna()  # Удаляем пустые строки
            na_count = initial_count - len(df)
            if na_count > 0:
                print(f"  Удалено {na_count} строк с пропущенными значениями")
            
            # Проверяем, остались ли данные после удаления пропусков
            if len(df) == 0:
                print(f"  Внимание: после удаления пропусков файл пуст!")
                continue
            
            # Добавляем столбцы из имени файла
            df['sensor'] = sensor_val
            df['source_file'] = source_file_val
            df['frequency'] = frequency_val
            df['boost'] = boost_val
            
            # Преобразуем типы данных
            df['sensor'] = df['sensor'].astype(str)
            df['source_file'] = df['source_file'].astype(str)
            df['frequency'] = df['frequency'].astype(str)
            df['boost'] = df['boost'].astype(str)
            
            # Форматируем acceleration
            df['acceleration'] = df['acceleration'].astype(str).str.replace('.', ',', regex=False)
            
            # Преобразуем acceleration в число для фильтрации
            df['acceleration_num'] = pd.to_numeric(df['acceleration'].str.replace(',', '.'), errors='coerce')
            
            # Удаляем строки с некорректными числовыми значениями
            invalid_num_count = df['acceleration_num'].isna().sum()
            if invalid_num_count > 0:
                df = df.dropna(subset=['acceleration_num'])
            
            # Удаляем строки с нулевыми значениями acceleration
            initial_count = len(df)
            df = df[df['acceleration_num'] != 0]
            zero_count = initial_count - len(df)
            if zero_count > 0:
                print(f"  Удалено {zero_count} строк с нулевыми значениями ускорения")
            
            # Проверяем, остались ли данные после фильтрации
            if len(df) == 0:
                print(f"  Внимание: после фильтрации файл пуст!")
                continue
            
            # УДАЛЕНИЕ ВЫБРОСОВ ПО IQR
            if len(df) > 0:
                Q1 = df['acceleration_num'].quantile(0.25)
                Q3 = df['acceleration_num'].quantile(0.75)
                IQR = Q3 - Q1
                lower_bound = Q1 - 1.5 * IQR
                upper_bound = Q3 + 1.5 * IQR
                
                # Фильтрация выбросов
                df_clean = df[(df['acceleration_num'] >= lower_bound) & 
                             (df['acceleration_num'] <= upper_bound)].copy()
                
                outliers_count = len(df) - len(df_clean)
                if outliers_count > 0:
                    print(f"  Удалено {outliers_count} выбросов по методу IQR")
                
                # Создаем строку со средним значением
                if len(df_clean) > 0:
                    mean_acceleration = df_clean['acceleration_num'].mean()
                    mean_row = pd.DataFrame({
                        'acceleration': [f"{mean_acceleration:.10f}".replace('.', ',')],
                        'frequency': [df_clean['frequency'].iloc[0]],
                        'boost': [df_clean['boost'].iloc[0]],
                        'sensor': [df_clean['sensor'].iloc[0]],
                        'source_file': [df_clean['source_file'].iloc[0]],
                        'axis': [axis],  # Добавляем информацию об оси
                        'source_filename': [filename]  # Сохраняем оригинальное имя файла для отладки
                    })
                    all_mean_rows.append(mean_row)
                    processed_files_count += 1
                    print(f"  Успешно обработан файл {filename} (sensor={sensor_val}). Среднее ускорение: {mean_acceleration:.6f}")
                    
        except Exception as e:
            print(f"Ошибка при обработке файла {filename}: {str(e)}")
            import traceback
            traceback.print_exc()
    
    # Сохраняем все строки со средними значениями
    if all_mean_rows:
        final_df = pd.concat(all_mean_rows, ignore_index=True).dropna()
        
        # Переупорядочиваем столбцы
        column_order = ['acceleration', 'frequency', 'boost', 'sensor', 'source_file', 'axis', 'source_filename']
        final_df = final_df[column_order]
        
        output_path = os.path.join(folder_path, "final.csv")
        final_df.to_csv(output_path, sep=';', index=False, encoding='cp1251')
        print(f"\nИтоговый файл сохранен: {output_path}")
        print(f"Обработано файлов: {processed_files_count}/{len(csv_files)}")
        print(f"Всего строк в final.csv: {len(final_df)}")
        
        # ========== 3. ДОБАВЛЕНИЕ СТОЛБЦА ОШИБКИ ==========
        def convert_to_float(value):
            if isinstance(value, str):
                return float(value.replace(',', '.'))
            return float(value)
        
        df = final_df.copy()
        df['acceleration_num'] = df['acceleration'].apply(convert_to_float)
        df['boost_num'] = df['boost'].apply(convert_to_float)
        df['error_%'] = abs(df['boost_num'] - df['acceleration_num']) / 5.6 * 100
        df['error_%'] = df['error_%'].round(2).astype(str).str.replace('.', ',')
        
        # Удаляем временные числовые столбцы
        df = df.drop(['acceleration_num', 'boost_num'], axis=1)
        
        # Сохраняем промежуточный результат
        df.to_csv(output_path, sep=';', index=False, encoding='cp1251')
        print(f"Добавлен столбец error_% для оси {axis}")
        
        # ========== АНАЛИЗ ПО СЕНСОРАМ (1 И 2) ==========
        sensor_values = df['sensor'].unique()
        print(f"\nНайдены сенсоры: {sensor_values}")
        
        # Создаем копию исходного DataFrame для добавления новых столбцов
        df_with_new_columns = df.copy()
        
        # Анализируем каждый сенсор отдельно
        for sensor_val in sorted(sensor_values):
            sensor_df = df_with_new_columns[df_with_new_columns['sensor'] == sensor_val].copy()
            print(f"\n{'─'*40}")
            print(f"АНАЛИЗ СЕНСОРА {sensor_val} (ОСЬ {axis})")
            print(f"{'─'*40}")
            print(f"Количество записей: {len(sensor_df)}")
            
            if len(sensor_df) < 2:
                print(f"  Слишком мало данных для анализа сенсора {sensor_val}")
                continue
            
            # ========== 4. ПОСТРОЕНИЕ МОДЕЛИ ДЛЯ OPTIMAL_K (для каждого сенсора) ==========
            # Преобразуем в числа для анализа
            df_numeric = sensor_df.copy()
            for col in ['acceleration', 'frequency', 'boost', 'error_%']:
                df_numeric[col] = df_numeric[col].astype(str).str.replace(',', '.').astype(float)
            
            # Находим оптимальные k для каждого frequency
            freq_values = sorted(df_numeric['frequency'].unique())
            optimal_k_by_freq = {}
            
            print(f"\nНахождение оптимальных k для каждого frequency (сенсор {sensor_val}, ось {axis}):")
            for freq_val in freq_values:
                freq_df = df_numeric[df_numeric['frequency'] == freq_val]
                if len(freq_df) == 0:
                    continue
                    
                k_values = np.linspace(0.95, 1.05, 1001)
                best_k = 1.0
                min_max_error = float('inf')
                
                for k in k_values:
                    errors = abs(freq_df['boost'] - k * freq_df['acceleration']) / 5.6 * 100
                    max_error = errors.max()
                    if max_error < min_max_error:
                        min_max_error = max_error
                        best_k = k
                
                optimal_k_by_freq[freq_val] = best_k
                print(f"  Frequency = {freq_val}: Optimal k = {best_k:.6f}")
            
            if not optimal_k_by_freq:
                print(f"  Не удалось найти оптимальные k для сенсора {sensor_val}")
                continue
            
            # Подготавливаем данные для регрессии
            freq_list = np.array(list(optimal_k_by_freq.keys()))
            k_list = np.array(list(optimal_k_by_freq.values()))
            
            # ========== 5. ГРАФИКИ РАЗЛИЧНЫХ МОДЕЛЕЙ ЗАВИСИМОСТИ K ОТ FREQUENCY ==========
            print(f"\nПостроение различных моделей зависимости k от frequency (сенсор {sensor_val}, ось {axis})...")
            fig = plt.figure(figsize=(15, 10))
            
            # 5.1 Кубическая модель
            ax1 = plt.subplot(2, 3, 1)
            poly3 = PolynomialFeatures(degree=3)
            X_freq = freq_list.reshape(-1, 1)
            X_poly3 = poly3.fit_transform(X_freq)
            model_poly3 = LinearRegression()
            model_poly3.fit(X_poly3, k_list)
            
            x_fit = np.linspace(min(freq_list)-2, max(freq_list)+5, 100).reshape(-1, 1)
            x_fit_poly3 = poly3.transform(x_fit)
            ax1.scatter(freq_list, k_list, s=80, color='blue', alpha=0.7, label='Оптимальные k')
            ax1.plot(x_fit, model_poly3.predict(x_fit_poly3), 'm-', linewidth=2, 
                    label='Кубическая (3 степень)')
            ax1.set_xlabel('Frequency')
            ax1.set_ylabel('Optimal k')
            ax1.set_title(f'Кубическая зависимость (сенсор {sensor_val}, ось {axis})')
            ax1.legend(fontsize=9)
            ax1.grid(True, alpha=0.3)
            
            # 5.2 Сравнение всех моделей (включая логарифмическую)
            ax2 = plt.subplot(2, 3, 2)
            
            # Линейная модель
            model_linear = LinearRegression()
            model_linear.fit(X_freq, k_list)
            
            # Параболическая модель
            poly2 = PolynomialFeatures(degree=2)
            X_poly2 = poly2.fit_transform(X_freq)
            model_poly2 = LinearRegression()
            model_poly2.fit(X_poly2, k_list)
            
            # Логарифмическая модель
            X_log = np.log(X_freq + 1)  # +1 чтобы избежать log(0)
            model_log = LinearRegression()
            model_log.fit(X_log, k_list)
            
            x_fit_poly2 = poly2.transform(x_fit)
            x_fit_log = np.log(x_fit + 1)
            
            ax2.scatter(freq_list, k_list, s=100, color='black', alpha=0.8, label='Оптимальные k', zorder=5)
            
            models = [
                ('Линейная', model_linear.predict(x_fit), 'red'),
                ('Параболическая', model_poly2.predict(x_fit_poly2), 'green'),
                ('Кубическая', model_poly3.predict(x_fit_poly3), 'magenta'),
                ('Логарифмическая', model_log.predict(x_fit_log), 'cyan'),
            ]
            
            for name, y_pred, color in models:
                ax2.plot(x_fit, y_pred, color=color, linewidth=2, alpha=0.7, label=name)
            
            ax2.set_xlabel('Frequency')
            ax2.set_ylabel('Optimal k')
            ax2.set_title(f'Сравнение всех моделей (сенсор {sensor_val}, ось {axis})')
            ax2.legend(fontsize=8, loc='upper right')
            ax2.grid(True, alpha=0.3)
            
            # 5.3 Остатки моделей
            ax3 = plt.subplot(2, 3, 3)
            model_names = ['Линейная', 'Параболическая', 'Кубическая', 'Логарифмическая']
            model_functions = [
                lambda f: model_linear.predict([[f]])[0],
                lambda f: model_poly2.predict(poly2.transform([[f]]))[0],
                lambda f: model_poly3.predict(poly3.transform([[f]]))[0],
                lambda f: model_log.predict([[np.log(f + 1)]])[0],
            ]
            
            residuals_data = []
            for i, (name, func) in enumerate(zip(model_names, model_functions)):
                residuals = []
                for f_val, k_opt in zip(freq_list, k_list):
                    k_pred = func(f_val)
                    residuals.append(abs(k_opt - k_pred))
                residuals_data.append(residuals)
            
            bp = ax3.boxplot(residuals_data, labels=model_names, patch_artist=True)
            colors = ['lightcoral', 'lightgreen', 'lightpink', 'lightblue']
            for patch, color in zip(bp['boxes'], colors):
                patch.set_facecolor(color)
            
            ax3.set_ylabel('Абсолютная ошибка (k_opt - k_pred)')
            ax3.set_title(f'Ошибки моделей (остатки) (сенсор {sensor_val}, ось {axis})')
            ax3.grid(True, alpha=0.3)
            
            # 5.4 Параболическая модель
            ax4 = plt.subplot(2, 3, 4)
            ax4.scatter(freq_list, k_list, s=80, color='blue', alpha=0.7, label='Оптимальные k')
            ax4.plot(x_fit, model_poly2.predict(x_fit_poly2), 'g-', linewidth=2, 
                    label='Параболическая (2 степень)')
            ax4.set_xlabel('Frequency')
            ax4.set_ylabel('Optimal k')
            ax4.set_title(f'Параболическая зависимость (сенсор {sensor_val}, ось {axis})')
            ax4.legend(fontsize=9)
            ax4.grid(True, alpha=0.3)
            
            # 5.5 Линейная модель
            ax5 = plt.subplot(2, 3, 5)
            ax5.scatter(freq_list, k_list, s=80, color='blue', alpha=0.7, label='Оптимальные k')
            ax5.plot(x_fit, model_linear.predict(x_fit), 'r-', linewidth=2, 
                    label=f'Линейная: k={model_linear.coef_[0]:.6f}f+{model_linear.intercept_:.6f}')
            ax5.set_xlabel('Frequency')
            ax5.set_ylabel('Optimal k')
            ax5.set_title(f'Линейная зависимость (сенсор {sensor_val}, ось {axis})')
            ax5.legend(fontsize=9)
            ax5.grid(True, alpha=0.3)
            
            # 5.6 Логарифмическая модель
            ax6 = plt.subplot(2, 3, 6)
            ax6.scatter(freq_list, k_list, s=80, color='blue', alpha=0.7, label='Оптимальные k')
            ax6.plot(x_fit, model_log.predict(x_fit_log), 'c-', linewidth=2, 
                    label='Логарифмическая')
            ax6.set_xlabel('Frequency')
            ax6.set_ylabel('Optimal k')
            ax6.set_title(f'Логарифмическая зависимость (сенсор {sensor_val}, ось {axis})')
            ax6.legend(fontsize=9)
            ax6.grid(True, alpha=0.3)
            
            plt.suptitle(f'Зависимость оптимального коэффициента k от frequency (сенсор {sensor_val}, ось {axis})', fontsize=16, y=0.98)
            plt.tight_layout()
            plt.show()
            
            # ========== 6. ПРОВЕРКА МОДЕЛЕЙ НА ВСЕМ ДАТАСЕТЕ ==========
            print(f"\nПроверка моделей на всем датасете (сенсор {sensor_val}, ось {axis})...")
            
            fig2 = plt.figure(figsize=(18, 6))
            
            # Функции для вычисления k по моделям
            def k_linear_model(f):
                return model_linear.predict([[f]])[0]
            
            def k_parabola_model(f):
                return model_poly2.predict(poly2.transform([[f]]))[0]
            
            def k_cubic_model(f):
                return model_poly3.predict(poly3.transform([[f]]))[0]
            
            def k_log_model(f):
                return model_log.predict([[np.log(f + 1)]])[0]
            
            # 6.1 Ошибки для кубической модели
            ax1 = plt.subplot(1, 3, 1)
            errors_cubic = []
            for idx, row in df_numeric.iterrows():
                freq_val = row['frequency']
                boost_val = row['boost']
                acc_val = row['acceleration']
                
                k = k_cubic_model(freq_val)
                error = abs(boost_val - k * acc_val) / 5.6 * 100
                errors_cubic.append(error)
            
            errors_cubic = np.array(errors_cubic)
            ax1.hist(errors_cubic, bins=20, alpha=0.7, color='magenta', edgecolor='black')
            ax1.axvline(x=1.5, color='black', linestyle='--', linewidth=2, label='1.5% порог')
            ax1.axvline(x=errors_cubic.max(), color='purple', linestyle='-', linewidth=1, 
                       label=f'Макс: {errors_cubic.max():.3f}%')
            ax1.set_xlabel('Ошибка, %')
            ax1.set_ylabel('Частота')
            ax1.set_title(f'Кубическая модель (сенсор {sensor_val}, ось {axis})\nСредняя: {errors_cubic.mean():.3f}%')
            ax1.legend(fontsize=9)
            ax1.grid(True, alpha=0.3)
            
            # 6.2 Сравнение всех моделей (boxplot)
            ax2 = plt.subplot(1, 3, 2)
            
            # Вычисляем ошибки для всех моделей
            errors_linear = []
            errors_parabola = []
            errors_log = []
            
            for idx, row in df_numeric.iterrows():
                freq_val = row['frequency']
                boost_val = row['boost']
                acc_val = row['acceleration']
                
                # Линейная модель
                k = k_linear_model(freq_val)
                error = abs(boost_val - k * acc_val) / 5.6 * 100
                errors_linear.append(error)
                
                # Параболическая модель
                k = k_parabola_model(freq_val)
                error = abs(boost_val - k * acc_val) / 5.6 * 100
                errors_parabola.append(error)
                
                # Логарифмическая модель
                k = k_log_model(freq_val)
                error = abs(boost_val - k * acc_val) / 5.6 * 100
                errors_log.append(error)
            
            all_errors = [errors_linear, errors_parabola, errors_cubic, errors_log]
            model_names_box = ['Линейная', 'Параболическая', 'Кубическая', 'Логарифмическая']
            colors_box = ['lightcoral', 'lightgreen', 'lightpink', 'lightblue']
            
            bp = ax2.boxplot(all_errors, labels=model_names_box, patch_artist=True)
            for patch, color in zip(bp['boxes'], colors_box):
                patch.set_facecolor(color)
            
            ax2.axhline(y=1.5, color='black', linestyle='--', linewidth=2, label='1.5% порог')
            ax2.set_ylabel('Ошибка, %')
            ax2.set_title(f'Сравнение распределений ошибок (сенсор {sensor_val}, ось {axis})')
            ax2.legend(fontsize=9)
            ax2.grid(True, alpha=0.3)
            
            # 6.3 Процент строк с ошибкой ≤ 1.5%
            ax3 = plt.subplot(1, 3, 3)
            below_threshold = [
                (np.array(errors_linear) <= 1.5).sum() / len(errors_linear) * 100,
                (np.array(errors_parabola) <= 1.5).sum() / len(errors_parabola) * 100,
                (errors_cubic <= 1.5).sum() / len(errors_cubic) * 100,
                (np.array(errors_log) <= 1.5).sum() / len(errors_log) * 100
            ]
            
            bars = ax3.bar(model_names_box, below_threshold, color=['red', 'green', 'magenta', 'cyan'], alpha=0.7)
            ax3.axhline(y=100, color='green', linestyle='--', linewidth=1, alpha=0.5, label='100%')
            ax3.set_xlabel('Модель')
            ax3.set_ylabel('Процент строк с error ≤ 1.5%')
            ax3.set_title(f'Эффективность моделей (сенсор {sensor_val}, ось {axis})')
            ax3.set_ylim([0, 105])
            
            # Добавляем значения на столбцы
            for bar, value in zip(bars, below_threshold):
                height = bar.get_height()
                ax3.text(bar.get_x() + bar.get_width()/2., height + 1,
                        f'{value:.1f}%', ha='center', va='bottom', fontsize=9)
            
            ax3.legend(fontsize=9)
            ax3.grid(True, alpha=0.3)
            
            plt.suptitle(f'Результаты применения моделей на всем датасете (сенсор {sensor_val}, ось {axis})', fontsize=16, y=0.98)
            plt.tight_layout()
            plt.show()
            
            # ========== 7. ВЫБОР ЛУЧШЕЙ МОДЕЛИ И ДОБАВЛЕНИЕ В ФАЙЛ ==========
            # Определяем лучшую модель по минимальной максимальной ошибке
            model_errors_list = [errors_linear, errors_parabola, errors_cubic, errors_log]
            model_max_errors = [max(err) for err in model_errors_list]
            best_model_idx = np.argmin(model_max_errors)
            best_model_name = model_names_box[best_model_idx]
            
            print(f"\nРезультаты моделей (сенсор {sensor_val}, ось {axis}):")
            print(f"{'Модель':<15} {'Макс. ошибка':<12} {'Сред. ошибка':<12} {'≤1.5% строк':<12}")
            print("-" * 60)
            
            for i, name in enumerate(model_names_box):
                max_err = max(model_errors_list[i])
                mean_err = np.mean(model_errors_list[i])
                count_below = (np.array(model_errors_list[i]) <= 1.5).sum()
                percent_below = count_below / len(df_numeric) * 100
                
                print(f"{name:<15} {max_err:<12.6f} {mean_err:<12.6f} {percent_below:<12.1f}%")
            
            print(f"\nЛучшая модель для сенсора {sensor_val} (ось {axis}): {best_model_name}")
            print(f"Максимальная ошибка: {model_max_errors[best_model_idx]:.6f}%")
            
            # Используем кубическую модель (как лучшую)
            def k_cubic_formula(frequency):
                X_input = poly3.transform([[frequency]])
                return model_poly3.predict(X_input)[0]
            
            # Добавляем столбцы в DataFrame сенсора
            sensor_df['Optimal_k'] = df_numeric['frequency'].apply(k_cubic_formula)
            sensor_df['new_error_%'] = abs(df_numeric['boost'] - sensor_df['Optimal_k'] * df_numeric['acceleration']) / 5.6 * 100
            
            # Форматируем числа
            for col in ['Optimal_k', 'new_error_%']:
                sensor_df[col] = sensor_df[col].apply(lambda x: f"{float(x):.10f}".replace('.', ',') if pd.notnull(x) else x)
            
            # Выводим уравнение кубической модели
            coefficients = model_poly3.coef_
            intercept = model_poly3.intercept_
            print(f"\nУравнение кубической модели для сенсора {sensor_val} (ось {axis}):")
            print(f"k = {coefficients[3]:.10f}·f³ + {coefficients[2]:.10f}·f² + {coefficients[1]:.10f}·f + {intercept:.10f}")
            
            # Обновляем исходный DataFrame правильным способом - по индексам
            for idx in sensor_df.index:
                if idx in df_with_new_columns.index:
                    df_with_new_columns.at[idx, 'Optimal_k'] = sensor_df.at[idx, 'Optimal_k']
                    df_with_new_columns.at[idx, 'new_error_%'] = sensor_df.at[idx, 'new_error_%']
            
            # ========== 8. АНАЛИЗ И ВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ ==========
            # Преобразуем все числовые столбцы для анализа
            for col in sensor_df.columns:
                if col not in ['sensor', 'source_file', 'axis']:
                    try:
                        sensor_df[col] = sensor_df[col].astype(str).str.replace(',', '.').astype(float)
                    except:
                        pass
            
            # Статистика ошибок
            old_errors = sensor_df['error_%'].values
            new_errors = sensor_df['new_error_%'].values
            
            stats_old = {
                'max': float(old_errors.max()),
                'mean': float(old_errors.mean()),
                'median': float(np.median(old_errors)),
                'below_15': int((old_errors <= 1.5).sum()),
                'total': len(old_errors)
            }
            
            stats_new = {
                'max': float(new_errors.max()),
                'mean': float(new_errors.mean()),
                'median': float(np.median(new_errors)),
                'below_15': int((new_errors <= 1.5).sum()),
                'total': len(new_errors)
            }
            
            # Основной график сравнения
            plt.figure(figsize=(15, 5))
            
            # 1. Гистограммы
            plt.subplot(1, 3, 1)
            bins = 30
            max_error_val = max(stats_old['max'], stats_new['max'])
            bins_range = np.linspace(0, max_error_val * 1.1, bins + 1)
            
            plt.hist(old_errors, bins=bins_range, alpha=0.6, color='red', 
                     label=f'Старые (макс: {stats_old["max"]:.2f}%)', edgecolor='darkred')
            plt.hist(new_errors, bins=bins_range, alpha=0.6, color='green', 
                     label=f'Новые (макс: {stats_new["max"]:.2f}%)', edgecolor='darkgreen')
            plt.axvline(x=1.5, color='black', linestyle='--', linewidth=2, label='Порог 1.5%')
            plt.xlabel('Ошибка, %')
            plt.ylabel('Количество наблюдений')
            plt.title(f'Распределение старых и новых ошибок\n(сенсор {sensor_val}, ось {axis})')
            plt.legend()
            plt.grid(True, alpha=0.3)
            
            # 2. Boxplot
            plt.subplot(1, 3, 2)
            boxplot_data = [old_errors, new_errors]
            labels = [f'Старые\nмакс: {stats_old["max"]:.2f}%', f'Новые\nмакс: {stats_new["max"]:.2f}%']
            
            bp = plt.boxplot(boxplot_data, labels=labels, patch_artist=True)
            colors_box = ['lightcoral', 'lightgreen']
            for i, box in enumerate(bp['boxes']):
                box.set(facecolor=colors_box[i], alpha=0.7)
            plt.axhline(y=1.5, color='black', linestyle='--', linewidth=2)
            plt.ylabel('Ошибка, %')
            plt.title(f'Boxplot распределений ошибок\n(сенсор {sensor_val}, ось {axis})')
            plt.grid(True, alpha=0.3, axis='y')
            
            # 3. Кумулятивное распределение
            plt.subplot(1, 3, 3)
            sorted_old = np.sort(old_errors)
            sorted_new = np.sort(new_errors)
            cumulative_old = np.arange(1, len(sorted_old) + 1) / len(sorted_old) * 100
            cumulative_new = np.arange(1, len(sorted_new) + 1) / len(sorted_new) * 100
            
            plt.plot(sorted_old, cumulative_old, 'r-', linewidth=2, label='Старые ошибки')
            plt.plot(sorted_new, cumulative_new, 'g-', linewidth=2, label='Новые ошибки')
            plt.axvline(x=1.5, color='black', linestyle='--', linewidth=2)
            plt.xlabel('Ошибка, %')
            plt.ylabel('Процент строк с ошибкой ≤ X, %')
            plt.title(f'Кумулятивное распределение\n(сенсор {sensor_val}, ось {axis})')
            plt.legend()
            plt.grid(True, alpha=0.3)
            plt.ylim([0, 105])
            
            plt.suptitle(f'Сравнение ошибок до и после корректировки (сенсор {sensor_val}, ось {axis})', fontsize=14, fontweight='bold')
            plt.tight_layout()
            plt.show()
            
            # Вывод результатов
            print(f"\nРЕЗЮМЕ АНАЛИЗА ДЛЯ СЕНСОРА {sensor_val} (ОСЬ {axis})")
            print("="*60)
            print(f"\nСтарые ошибки:")
            print(f"  Максимальная: {stats_old['max']:.4f}%")
            print(f"  Средняя: {stats_old['mean']:.4f}%")
            print(f"  ≤1.5%: {stats_old['below_15']}/{stats_old['total']} ({stats_old['below_15']/stats_old['total']*100:.1f}%)")
            
            print(f"\nНовые ошибки:")
            print(f"  Максимальная: {stats_new['max']:.4f}%")
            print(f"  Средняя: {stats_new['mean']:.4f}%")
            print(f"  ≤1.5%: {stats_new['below_15']}/{stats_new['total']} ({stats_new['below_15']/stats_new['total']*100:.1f}%)")
            
            print(f"\nУлучшение для сенсора {sensor_val} (ось {axis}):")
            print(f"  Снижение максимальной ошибки: {stats_old['max'] - stats_new['max']:.4f}%")
            print(f"  Увеличение строк ≤1.5%: +{stats_new['below_15'] - stats_old['below_15']} строк")
            print("="*60)
        
        # Теперь используем df_with_new_columns вместо df
        df = df_with_new_columns.copy()
        
        # Сохраняем обновленный файл для оси с новыми столбцами
        df.to_csv(output_path, sep=';', index=False, encoding='cp1251')
        print(f"\nФайл обновлен с Optimal_k и new_error_%: {output_path}")
        
        # Добавляем обработанные данные в общий список
        all_sensors_data.append(df)
        
    else:
        print(f"\nНе удалось обработать ни один файл для оси {axis}!")

# ========== СОХРАНЕНИЕ ВСЕХ ДАННЫХ В ОДИН ФАЙЛ ==========
if all_sensors_data:
    print(f"\n{'='*80}")
    print("СОХРАНЕНИЕ ВСЕХ ДАННЫХ В ОДИН ОБЩИЙ ФАЙЛ")
    print(f"{'='*80}")
    
    # Объединяем все данные
    all_data_df = pd.concat(all_sensors_data, ignore_index=True)
    
    # Проверяем наличие столбцов Optimal_k и new_error_%
    print(f"\nПроверка наличия новых столбцов:")
    print(f"  Optimal_k присутствует: {'Optimal_k' in all_data_df.columns}")
    print(f"  new_error_% присутствует: {'new_error_%' in all_data_df.columns}")
    print(f"  Все столбцы: {list(all_data_df.columns)}")
    
    # Если столбцы отсутствуют, создаем их с NaN значениями
    if 'Optimal_k' not in all_data_df.columns:
        print("  Создаем столбец Optimal_k с NaN значениями")
        all_data_df['Optimal_k'] = np.nan
    
    if 'new_error_%' not in all_data_df.columns:
        print("  Создаем столбец new_error_% с NaN значениями")
        all_data_df['new_error_%'] = np.nan
    
    # Переупорядочиваем столбцы для лучшей читаемости
    column_order = ['acceleration', 'frequency', 'boost', 'sensor', 'source_file', 'axis', 
                   'Optimal_k', 'error_%', 'new_error_%', 'source_filename']
    
    # Оставляем только существующие столбцы
    existing_columns = [col for col in column_order if col in all_data_df.columns]
    # Добавляем остальные столбцы, если они есть
    other_columns = [col for col in all_data_df.columns if col not in existing_columns]
    final_column_order = existing_columns + other_columns
    
    all_data_df = all_data_df[final_column_order]
    
    # Сортируем по оси, сенсору и частоте
    all_data_df = all_data_df.sort_values(['axis', 'sensor', 'frequency', 'boost'])
    
    # Сохраняем в общий файл
    all_data_df.to_csv(final_all_path, sep=';', index=False, encoding='cp1251')
    
    print(f"\nВсе данные сохранены в: {final_all_path}")
    print(f"Общее количество строк: {len(all_data_df)}")
    print(f"Размер файла: {os.path.getsize(final_all_path) / 1024:.2f} KB")
    
    # Проверяем наличие новых столбцов в сохраненном файле
    print(f"\nПроверка сохраненного файла:")
    try:
        saved_df = pd.read_csv(final_all_path, delimiter=';', encoding='cp1251')
        print(f"  Загружено строк: {len(saved_df)}")
        print(f"  Столбцы в файле: {list(saved_df.columns)}")
        
        if 'Optimal_k' in saved_df.columns:
            non_null_count = saved_df['Optimal_k'].notna().sum()
            print(f"  Optimal_k: {non_null_count} непустых значений")
        
        if 'new_error_%' in saved_df.columns:
            non_null_count = saved_df['new_error_%'].notna().sum()
            print(f"  new_error_%: {non_null_count} непустых значений")
    except Exception as e:
        print(f"  Ошибка при чтении сохраненного файла: {e}")

print("\n" + "="*80)
print("ОБРАБОТКА ВСЕХ ОСЕЙ И СЕНСОРОВ ЗАВЕРШЕНА!")
print("="*80)
```
